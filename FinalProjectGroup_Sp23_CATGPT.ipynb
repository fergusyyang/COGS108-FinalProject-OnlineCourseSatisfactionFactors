## Background & Prior Work

Online learning has revolutionized access to education, providing people worldwide with opportunities to learn about various topics. As an evolving learning method that relies on platform and technology, online courses require constant revisions to fit with students’ needs. However, instructional designers may sometimes overlook non-teaching factors when developing courses, which can result in lower student satisfaction rates and enrollment [^Yu]. To bridge the gap between designers and users and enhance the learning experience, it is essential to analyze large data sets covering various aspects of online courses. 


The use of educational data to inform online learning experience is not a novel concept. In 2006, the study by Romero and Ventura discussed how statistical analysis such as regression could be applied to learning time and number of attempts to design effective educational systems [^Ventura and Romero]. Additionally, text mining was also used to evaluate specific knowledge grasp by students through online discussion [^Ventura and Romero]. This study points out how educational data can be used to measure learning outcomes and satisfaction in web-based courses. In 2008, another study utilized structural equation analysis, a multivariate analysis, to investigate the relationship between student’s perceptions on their computer ability, usefulness and flexibility in online courses, and their satisfaction with remote learning [^Sahin and Shelley]. The study found that all the three factors on students’ perceptions were positively correlated with student satisfaction in distance learning. 

Building upon previous research on students’ attitudes about online learning, recent studies focus on more accurate predictors in course design to improve students’ online learning experience. A study found that module designs manage course workload and flexibility, such as assessment frequency, duration of course, number of classes, level of difficulty, are all related to learning satisfaction [^Li, Marsh and Rienties]. In this study, overall student satisfaction data is collected, and the degree of influence for each block of predictors was measured by logistic regression analysis. Further study provided evidence for module difficulty being an important factor influencing students’ satisfaction in online learning [^Sutherland, Warwick and Anderson]. Similarly, through a regression analysis on survey data, another study found that non-teaching factors such as price and class size also affect students’ online learning experience, while teacher credentials are not a significant predictor [^Wart, Ni, Ready,Shayo, and Court].

**Hypothesis test**

To investigate the relationship between average review points and course difficulty level, we hope to conduct a hypothesis test to verify if there is a statistically significant difference between review points that can be accounted for by course fifficulty level. Our hypotheses are:

- Null Hypothesis (H0): The mean review scores are the same across all difficulty levels.
- Alternative Hypothesis (H1): At least one difficulty level has a different mean review score.

We plan to perform a one-way ANOVA test on the two variables. Before using an ANOVA test, it is important to check if our data meet the following assumptions for ANOVA:

- Randomness and Independence
- Assumption of Normality
- Homogeneity of Variance
- Dependent Variable Is Continuous

### 1. Assumption of Independence - satisfied
Since the dataset includes almost 5000 dataset and due to the nature of the dataset, it is unlikely for one observation selection to impact the selection of another observation. So we feel safe to assume that each data is independent from each other

```python
sample = online_course[["reviews_avg", "level"]]
```

In this way, each observation is randomly selected and thus independent from other observations.

### 2. Assumption of Normality - satisfied
To conduct ANOVA test, we need to make sure our data is normally distributed. We drew out the histogram for each level to look for normality. Due to the sample size, the graphs below don't show a robust normality. However, the distributions for each level are seem to be roughly normally distributed. We will take a note of this when analysis for the limitation. But for now, we will work on checking the rest of the conditions for anova test.

```python
import numpy as np 
import scipy.stats as stats
import matplotlib.pyplot as plt

unique_level = online_course['level'].unique()
for l in unique_level:
    plt.hist(online_course[online_course['level']==l]['reviews_avg'], bins=10, edgecolor='black')
    plt.title(" Plot for " +  l)
    plt.show()
```

### 3. Homogeneity of Variance -- satisfied
To ensure that variances of dependent variable are roughly equal between difficulty level, we calculate a ratio of the largest to the smallest sample standard deviation for dependent variable reviews_avg. We also conduct levene_test to check whether variance across different levels are the same. The result shows that we fail to reject null hypothesis, the variance across different levels are the same.

```python
std_devs = sample.groupby('level')['reviews_avg'].std()
ratio = std_devs.max() / std_devs.min()
print("\nRatio of largest to smallest standard deviation: ", ratio)

levene_test = levene(*[group['reviews_avg'].values for name, group in sample.groupby('level')])
print(f"Levene's Test: Test-statistic = {levene_test.statistic}, p-value = {levene_test.pvalue}")
if levene_test.pvalue <0.05:
    print ("we reject null hypothesis")
else:
    print ("we fail to reject null hypothesis, the variance across different levels are the same")
```

### 4. Dependent Variable Is Continuous - satisfied
We want to know how course difficulty may influence average review, so average review is the dependent variable. It contains scores from 0 to 5 in a float type, as to be checked as following:

```python
sample.dtypes
```

The assumption for the dependent variable to be continuous is satisfied.

### One-way ANOVA Test
Finally, we conducted the one-way ANOVA test to examine if there is a statistically significant difference in average review scores across different course difficulty levels.

**H0: there is no difference in means for courses' review rate across different level.**

**H1: there is a difference in means for courses' review rate across different level.**

```python

print("\nOne-way ANOVA Test:")
anova_result = stats.f_oneway(sample['reviews_avg'][sample['level'] == 'All Levels'],
               sample['reviews_avg'][sample['level'] == 'Beginner'],
               sample['reviews_avg'][sample['level'] == 'Intermediate'],
               sample['reviews_avg'][sample['level'] == 'Expert'])
print(f"F-statistic = {anova_result.statistic}, p-value = {anova_result.pvalue}")
```

```python
model = ols('reviews_avg ~ C(level)', data=sample).fit()
aov_table = sm.stats.anova_lm(model, typ=2)
aov_table
```

```python
if anova_result.pvalue < 0.05:
    print("\nThere is a statistically significant difference in average review points that can be attributed to difficulty levels. We reject the null hypothesis.")
else:
    print("\nThere is no statistically significant difference in average review points that can be attributed to difficulty levels. We cannot reject the null hypothesis.")
```

### Kruskal-Wallis H-test
As our graph showed below, the normality condition is not robustly satisfied. Therefore we will conduct a Kruskal-Wallis H-test to check whether the medians of the difficulty level groups are significantly different. In this section, we will still compare between the four different levels. We hope to use this test to help us understand whether the difficulty level would affect the course ratings.

**H0: there is no difference in medians for courses' review rate across four differnt levels.**

**H1: there is a difference in medians for courses' review rate across four differnt levels.**

```python

sample1 = sample[sample['level']=='All Levels']['reviews_avg']
sample2 = sample[sample['level']=='Beginner']['reviews_avg']
sample3 = sample[sample['level']=='Intermediate']['reviews_avg']
sample4 = sample[sample['level']=='Expert']['reviews_avg']


statistic, p_value = stats.kruskal(sample1, sample2, sample3,sample4)
print("Kruskal-Wallis H statistic:", statistic)
print("p-value:", p_value)
```

Since p-value is smaller than 0.05, we reject null hypothesis. We have evidence to say that there is a difference in medians for courses' review rate across four differnt levels. This corresponds to our result from anova test.

# Conclusion & Discussion

This project investigates factors influencing student ratings for online courses. Using regression analysis and hypothesis testing, it was found that course duration/lectures count has a positive correlation with average course ratings, and difficulty levels play a role. This work builds on previous research, aiming to bridge the gap between course designers and users, thereby enhancing the online learning experience.

The data at hand pertains to online course characteristics and user reviews. The goal is to determine the influence of various factors, such as course duration, lecture count, price, and difficulty level, on course review averages. The data encompasses a multitude of variables, including the average course review score ('reviews_avg'), price before ('main_price') and after discount ('price_after_discount'), course duration, lecture count, and price discount rate. Moreover, it includes a categorical variable representing the level of course difficulty. The distribution of average reviews is slightly left-skewed, with most courses receiving ratings around 4.5 and very few receiving average ratings around 2.0.

Our main question revolves around determining the relationships between the 'reviews_avg' variable and the other numerical variables: 'main_price', 'course_duration', 'lectures_count', and 'price_discount'. Initial analysis through scatter matrix plots and linear regression suggests potential correlations between these variables. For example, as 'course_duration' and 'lectures_count' seem to increase, 'reviews_avg' appears to increase as well. 'main_price' does not significantly affect the average review score, and 'price_discount' seems to have no strong linear relationship with 'reviews_avg'. We are also curious if the categorical variable 'level', representing course difficulty, affects review scores and plan to conduct further statistical tests to confirm these observations.

To determine the relationship between the proposed variables and review_avg, we conducted a linear regression. Noticed that 'course_duration' and 'lectures_count' are tend to cluster at the left side on our scatterplot, we decided to take the log of these two varibales for further anaylsis. It turns out that only 'course_duration_log' and 'lectures_count_log' have a more observable linear relationship between review__avg. The relationships between review_avg and "price" or "discount" were not as significant as we believed in the beginning. Since there's multi-conlinearity between course_duration_log and lectures_count_log, we will stay with single independent variable method for our linear regreassion anaylis. The linear regression indicates the relationship between the average review and the log of course duration can be linearly fitted as reviews_avg= 4.1514 + 0.18 * (course_duration_log). The linear regression indicates the relationship between the average review and the log of lectures count can be linearly fitted as reviews_avg= 3.8244 + 0.267 * lectures_count_log)

To analyze whether difficulty level would contribute to different course ratings. We first conducted an one-way anova test across four diffferent levels. The result shows that we should reject null hypothesis, which indicates that we have evidence to claim at least one of the course ratings across 4 levels is different. However, since the normality conditon has not been met perfectly. We followed up the analysis with Kruskal-Wallis H-test. This result also shows that we need to reject null hypothesis, which corresponds to our result from one-way anova. Therefore, we conclude that at least one of the course ratings across 4 levels is different. Course difficulty level does contribute to the course ratings.

Despite all the findings, our study has limitations on the scope of data and the types of analyses we’ve done. All the data was collected based on Udemy courses, so the characteristics and contents specific to Udemy courses may influence students’ review, which affect the external validity of our study. Future research could focus on a more diverse dataset with a wide range of courses pulled from different platforms and subjects to ensure more precise generalization. Due to time and the capacity of this study, we only focused on quantitative analysis using the average score students were given. Further research could be made based on qualitative text analysis to gain more comprehensive insights on students’ attitude toward the classes and inform a more detailed relationship between course design and students’ learning experience. 

The results of our study shed light on the important aspects in online courses that influence students’ learning experience and revealed users’ needs for e-learning systems. The findings can contribute to the design process and iterations of onlines classes and educational products, helping instructional designers to create more user-friendly online courses. With the potential of remote learning, a series of well-refined online courses could help make education more accessible to a wide range of student population. Therefore, this study has contributions to the development of educational technology  and education equity.
